### FILTERING

#### Filter Male and Female Passengers

```python
# FILTER BASED ON CERTAIN CONDITIONS
# GET NUMBER MALE PASSENGERS
male_passengers = raw_df.loc[raw_df.Sex == 'male',:]
print("Number of Male Passengers: {0} ".format(len(male_passengers)))
```

```python
#GET MALE PASSENGERS IN FIRST CLASS
# USE AND &, OR || IN CODE
male_passengers_1stclass = raw_df.loc[((raw_df.Sex == 'male')&(raw_df.Pclass == 1)),:]
print("Number of Male Passengers in 1st Class: {0} ".format(len(male_passengers_1stclass)))
```

```python
#GET NUMBER FEMALE PASSENGERS
female_passengers = raw_df.loc[raw_df.Sex == 'female',:]
print("Number of Female Passengers: {0}".format(len(female_passengers)))
```

```python
#GET NUMBER FEMALE PASSENGERS IN FIRST CLASS
female_passengers_1stclass = raw_df.loc[((raw_df.Sex == 'female')&(raw_df.Pclass == 1))]
print("Number of Female Passengers in 1st Class: {0}".format(len(female_passengers_1stclass)))
```

### numerical feature
### centrality measures

```python
#numerical feature
# centrality measures
print("mean fare: {0}".format(raw_df.Fare.mean()))#Mean
print("median fare: {0}".format(raw_df.Fare.median()))#Median
```

```python
#numerical feature
# centrality measures
print("mean age: {0}".format(raw_df.Age.mean()))#Mean
print("median age: {0}".format(raw_df.Age.median()))#Median
```

### Dispersion Measures

```python
#dispersion measures
print("min fare: {0}".format(raw_df.Fare.min()))#Minimum
print("max fare: {0}".format(raw_df.Fare.max()))#Maximum
print("fare range: {0}".format(raw_df.Fare.max() - raw_df.Fare.min()))#range
print("25th percentile {0}".format(raw_df.Fare.quantile(.25)))# 25th percentile
print("50th percentile {0}".format(raw_df.Fare.quantile(.50)))# 50th percentile
print("75th percentile {0}".format(raw_df.Fare.quantile(.75)))# 75th percentile
print("Variance fare: {0}".format(raw_df.Fare.var()))#Variance
print("Standard Deviation fare: {0}".format(raw_df.Fare.std()))# Standard Deviation

```

```python
# box - whisker plot
raw_df.Fare.plot(kind = 'box')
```
